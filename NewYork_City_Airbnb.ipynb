#Missing Data Handling


---


* Missing Data Handling: Dealing with missing values by either imputing them or making
informed decisions on how to handle gaps in the dataset.

```python
df['reviews_per_month'].isna()
```

```python
df.reviews_per_month.fillna(df.reviews_per_month.mean(),inplace=True)
df.head()
```

```python
df['last_review'].nunique()
```

```python
df.loc[df['last_review']<'2010-01-01']
```

#Standardization:


---


* Consistent formatting and units across the dataset
for accurate analysis.

```python
df.info()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_copy = df.copy()
cols = ['reviews_per_month','calculated_host_listings_count','availability_365']
df_copy[cols] = scaler.fit_transform(df_copy[cols])
```

```python
df.head()
```

```python
df_copy['calculated_host_listings_count'].mean()
df_copy['availability_365'].mean()
```

#Outlier Detection:


---


 * Identifying and addressing outliers that may skew analysis or model
performance.

```python
df.head(2)
```

```python
from scipy import stats
```

```python
z_scores = stats.zscore(df_copy[cols])
abs_z_score = np.abs(z_scores)
threshold = 3

outliers = (abs_z_score>threshold).any(axis = 1)
outlier_index = df_copy.index[outliers]
print("Indices of Outlier : ",outlier_index)
print("z_scores : ",z_scores)
print("Abs_z_score : ",abs_z_score)
```

**Remove Outlier**

```python
cleaned_df = df_copy[~outliers]
print("Original Dataset Length : ",len(df))
print("Cleaned Dataset Length : ",len(cleaned_df))
```

#Key Analysis



---


*  Here We First upload the data and then find which column has null data

*  Then we remove Duplicates and handle the null data

*  Then we Use StandardScaler to make our data on the same Scale.This will reduce the impact of outlier on our model

*  Then we Use z_score to find how many standard deviation a data point is from the mean and then we remove outlier to make our data more convinent to analyze
